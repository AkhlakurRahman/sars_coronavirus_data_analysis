## Importing necessary libraries

```python
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

## Querying target search

```python
target = new_client.target
target_query = target.search('coronavirus')
targets = pd.DataFrame.from_dict(target_query)
targets
```

## Select and retrieve Bioactivity data for SARS coronavirus 3c like proteinase

```python
selected_target = targets.target_chembl_id[4]
selected_target
```

Bioactivity data for Coronavirus 3c like proteinase that are reported as IC50 value

```python
activity = new_client.activity
res = activity.filter(target_chembl_id=selected_target).filter(standard_type='IC50')

df = pd.DataFrame.from_dict(res)
df
```

Creating a CSV file with the data above

```python
df.to_csv('bioactivity_data.csv', index=False)
```

If **standard_value** is missing from any compound, drop it

```python
df2 = df[df.standard_value.notna()]
df2
```

## Data preprocessing of bioactivity data

#### Labeling compound as either being active, inactive or intermediate

```python
bioactivity_class = []
for i in df2.standard_value:
    if float(i) >= 10000:
        bioactivity_class.append('inactive')
    elif float(i) <= 1000:
        bioactivity_class.append('active')
    else:
        bioactivity_class.append('intermediate')

bioactivity_class
```

## **Extracting *molecule_chembl_id* to a list**

```python
molecule_chembl_id = []
for i in df2.molecule_chembl_id:
    molecule_chembl_id.append(i)
```

## **Extracting *canonical_smiles* to a list**

```python
canonical_smiles = []
for i in df2.canonical_smiles:
    canonical_smiles.append(i)
```

## **Extracting *standard_value* to a list**

```python
standard_value = []
for i in df2.standard_value:
    standard_value.append(i)
```

## Combine 4 llists into a data frame

```python
data_tuple = list(zip(molecule_chembl_id, canonical_smiles, bioactivity_class, standard_value))
df3 = pd.DataFrame(data_tuple, columns=['molecule_chembl_id', 'canonical_smiles', 'bioactivity_class', 'standard_value'])
```

```python
df3
```

## Alternative (Efficient) method

```python
selected_columns = ['molecule_chembl_id', 'canonical_smiles', 'standard_value']
df3 = df2[selected_columns]
df3
```

```python
pd.concat([df3,pd.Series(bioactivity_class)], axis=1)
```
